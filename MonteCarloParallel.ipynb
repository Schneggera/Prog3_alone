```python
import random
import multiprocessing


def monte_carlo(run):
    '''berechnet Pi'''
    hits = i = 0
    while i < run:
        x = random.random()
        y = random.random()
        if x*x + y*y <= 1:
            hits += 1
        i += 1
    return(hits)


if __name__ == '__main__':
    runs = [100, 100, 100, 100]
    process_pool = multiprocessing.Pool(processes = 4)
    l2 = process_pool.map(monte_carlo, runs)
    print(4*sum(l2)/sum(runs))
```

```python
'''importiert Randomfunktion'''
import random
import multiprocessing
import cProfile

def monte_carlo(qIN, qOUT):
    ''' berechnet Pi '''
    while True:
        hits = 0
        for i in range(qIN.get()): #threading hier evtl sinnvoll?
            x = random.random()
            y = random.random()

            if (x**2 + y**2) <= 1:
                hits = hits + 1

        qOUT.put(hits)
        qIN.task_done()


if __name__ == '__main__':

    argumentQueue = multiprocessing.JoinableQueue()
    resultQueue = multiprocessing.Queue()
    nrOfProcesses = multiprocessing.cpu_count()
    processes = [multiprocessing.Process(target=monte_carlo, args=(argumentQueue, resultQueue))
                 for i in range(nrOfProcesses)]
    
    nr_per_call = 100000
    que_range = 10000

    for i in range(que_range):
        argumentQueue.put(nr_per_call)
    for p in processes:
        p.start()
    argumentQueue.join()
    for p in processes:
        p.terminate()

    summe = 0
    while not resultQueue.empty():
        summe = summe + resultQueue.get()
    print((summe/(que_range*nr_per_call))*4)
```
